```python
import os
import numpy as np
np.random.seed(42)

from scipy import sparse

import cytopath
import scvelo as scv
```

```python
# Import preprocessed dataset 
# https://github.com/velocyto-team/velocyto-notebooks/blob/master/python/DentateGyrus.ipynb
# Data was processed with the notebook above and then imported into the anndata format
# This version of the data can be download at 
# https://drive.google.com/file/d/1CJLkrdTehT33QjMBCnTvpcEq1Wy9WNtl/view

adata = cytopath.scvelo.read('DendateGyrus_velocyto.h5ad')
```

```python
# Visualise terminal states, biological annotaitons and louvain clustering
scv.tl.velocity_embedding(adata)
scv.pl.velocity_embedding_stream(adata, color=['root_cells', 'end_points','ClusterName'],
                                             ncols=3, color_map='YlGnBu')
```

```python
# Run cytopath
if sparse.issparse(adata.X):
    adata.X = adata.X.toarray()

cytopath.sampling(adata, cluster_key='ClusterName', num_cores=os.cpu_count()-1)
```

```python
# Infer trajectories
cytopath.trajectories(adata, basis='tsne', num_cores=os.cpu_count()-1)
```

```python
# Plot allignment and trajectories
cytopath.plot_trajectories(adata, directory='', basis='tsne', smoothing=False, size=10, figsize=(18,4))
```
